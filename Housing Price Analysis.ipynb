 ## LISTING PRICE ANALYSIS

```python
# Rename columns to eliminate intermediate spaces for querying

housing_data.rename(columns={"Median Listing Price":"MedianListingPrice","Median Sales Price":"MedianSalesPrice"},inplace=True)
housing_data


# Get data where Median Listing Price value is available
ListingPrice_data= housing_data.copy()
ListingPrice_data_df=ListingPrice_data[ListingPrice_data["MedianListingPrice"].notna()]
ListingPrice_data_df
```

```python
# Get the Top10 Regions based on high Listing Prices OVER THE YEARS FROM 2018 

high_LP=ListingPrice_data_df.sort_values("MedianListingPrice",ascending=False,ignore_index=True)
print(f"The top 10 regions with highest Listing Prices : \n{high_LP["RegionName"].head(10)}\n")
print(f"The lowest 10 regions  : \n{[high_LP["RegionName"].tail(10),high_LP["MedianListingPrice"].tail(10)]}  ")
```

```python
# Calculate the total Median Listing Price each Year

Total_LP=ListingPrice_data_df.groupby(["Year"])["MedianListingPrice"].sum()
LP_Yeardata=pd.DataFrame({
    "Yearly_LP_sum":Total_LP
}
                        )
YOY_LP_growth=LP_Yeardata["Yearly_LP_sum"].pct_change() *100
YOY_LP_growth_data =pd.DataFrame({
    "YOY_LP_growth":YOY_LP_growth
}
                                )   
YOY_LP_growth_data

# PLOT THE YOY GROWTH FOR MEDIAN LISTING PRICE

plt.plot(YOY_LP_growth_data, label="YOY_Growth")
plt.title('YOY_Percent_Growth') 
plt.xlabel('Year') 
plt.ylabel('YOY_Listing Price (%)') 
plt.legend() 
plt.tight_layout()
plt.show()
```

 ## SALES PRICE ANALYSIS

```python
SalesPrice_data= housing_data.copy()
SalesPrice_data_df=SalesPrice_data[SalesPrice_data["MedianSalesPrice"].notna()]
SalesPrice_data_df
```

```python
# GET TOP10 MEDIAN SALES PRICE OVER THE YEARS FROM 2018 

high_SP=SalesPrice_data_df.sort_values("MedianSalesPrice",ascending=False)
print(f"The top 10 regions with highest Sales Prices : \n{high_SP["RegionName"].head(10)}\n")
print(f"The lowest 10 regions  : \n{high_SP["RegionName"].tail(10)}  ")
high_SP.tail()
```

```python
# CALACULTE MEAN SALES PRICE IN A YEAR FOR A REGION

Avg_price_region= SalesPrice_data_df.groupby(['Year','RegionName'])["MedianSalesPrice"].mean()
Avg_price_region.reset_index()
Avg_price_region_df =pd.DataFrame({
    "MedianSalesPrice":Avg_price_region
}
                                )   

#TOP 10 REGIONS WITH HIGEST SALES PRICE

High_SP_byRegionYear=Avg_price_region_df.sort_values('MedianSalesPrice',ascending=False)
High_SP_byRegionYear=High_SP_byRegionYear.reset_index()
High_SP_byRegionYear
```

```python
# PLOT THE TOP 10 REGIONS WITH HIGHEST


plt.figure(figsize=(12, 6)) 
fig, ax = plt.subplots()
x=High_SP_byRegionYear["RegionName"].head(10)
y=High_SP_byRegionYear["MedianSalesPrice"].head(10)

ax.bar(x,y, label=x,linewidth=0.01)
# Augment or and incorporate the other graph properties
plt.title('HIGHEST SALES PRICE REGIONS (Top 10)') 
plt.xlabel('RegionName') 
plt.ylabel('MedianSalesPrice ($)') 
plt.xticks(rotation=90) 
plt.tight_layout() 
```

```python
# BOTTOM 10 REGIONS WITH LOWEST SALES PRICE

plt.figure(figsize=(12, 6)) 
fig, ax = plt.subplots()
x=High_SP_byRegionYear["RegionName"].tail(10)
y=High_SP_byRegionYear["MedianSalesPrice"].tail(10)

ax.bar(x, y, linewidth=0.01)
# Augment or and incorporate the other graph properties
plt.title(' LOWEST SALES PRICE REGIONS (Bottom 10)') 
plt.xlabel('RegionName') 
plt.ylabel('MedianSalesPrice ($)') 
plt.xticks(rotation=90) 
plt.tight_layout() 
```
